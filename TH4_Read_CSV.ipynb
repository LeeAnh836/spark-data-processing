Thực hiện viết bổ sung schema của file

```scala
import org.apache.spark.sql.types._

val schema = StructType(Array(
  StructField("age", IntegerType, true),
  StructField("job", StringType, true),
  StructField("marital", StringType, true),
  StructField("education", StringType, true),
  StructField("default", StringType, true),
  StructField("housing", StringType, true),
  StructField("loan", StringType, true),
  StructField("contact", StringType, true),
  StructField("month", StringType, true),
  StructField("dayOfWeek", StringType, true),
  StructField("duration", IntegerType, true),
  StructField("campaign", IntegerType, true),
  StructField("pdays", IntegerType, true),
  StructField("previous", IntegerType, true),
  StructField("poutcome", StringType, true),
  StructField("emVarRate", DoubleType, true),
  StructField("consPriceIdx", DoubleType, true),
  StructField("consConfIdx", DoubleType, true),
  StructField("euribor3m", DoubleType, true),
  StructField("nrEmployed", IntegerType, true),
  StructField("y", StringType, true)
))
```

Thực hiện read file csv. Lưu ý, cần truyền thêm các config cho phù hợp

```scala
val input = spark
  .read.format("csv")
//Thêm các config phù hợp vào đây
  .option("header", "true")
  .option("sep", ";")
  .schema(schema)
  .load(path)
```

```scala
import org.apache.spark.sql.functions._
```

Bài tập 1: Hãy kiểm tra xem có bao nhiêu người của mỗi nhóm tuổi. Trong đó, cách chia nhóm tuổi như sau: 
- `0-24` -> `Teen`
- `25-50` -> `Labor`
- `50 trở lên` -> `Retire`

Đầu ra mong muốn: `range_age`, `count`

```scala
val lesson1 = input.withColumn("range_age", when($"age" < 24, "Teen").when($"age" > 25 && $"age" < 50, "Labor").otherwise("Retire")).groupBy($"range_age").count
display(lesson1)
```

Bài tập 2: Thống kê số lượng người có nhà và không có nhà của từng nhóm tuổi. Đầu ra mong muốn: `range_age`, `house_status`, `count`

```scala
val lesson2 = input.withColumn("range_age", when($"age" < 24, "Teen").when($"age" > 25 && $"age" < 50, "Labor").otherwise("Retire"))
                  .withColumnRenamed("housing", "house_status")
                  .groupBy($"range_age", $"house_status").count
display(lesson2)
```

Bài tập 3: Thống kê những người đang có khoản vay. Theo nhóm tuổi và theo lượng người có nhà và không có nhà. Đầu ra mong muốn `range_age`, `house_status`, `loan_status`, `count`

```scala
val lesson3 = input.withColumn("range_age", when($"age" < 24, "Teen").when($"age" > 25 && $"age" < 50, "Labor").otherwise("Retire")).withColumnRenamed("housing", "house_status").groupBy($"range_age", $"house_status", $"loan").count
display(lesson3)
```
