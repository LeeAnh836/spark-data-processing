**Từ Câu 1 đến Câu 5, Thực hiện với danh sách đầu vào như sau:**

```scala
val input = List(1,2,3,4,5)
```

Câu 1: Hãy biến đối thành List mới bằng cách bình phương từng phần tử có trong danh sách

```scala
val lesson1 = input.map(x => x * x)
```

Câu 2: Hãy lọc các số lẻ và lập phương các số đó

```scala
val lesson2 = input.filter(x => x % 2 == 1).map(x => x * x * x)
```

Câu 3: Hãy tìm số trung bình trong dãy số

```scala
val lesson3 = input.sum.toDouble / input.length
```

Câu 4: Hãy tìm số trung vị trong dãy số

```scala
val lesson4 = input(input.length / 2)
```

Câu 5: Hãy tìm số nguyên lớn thứ 2 trong danh sách

```scala
val lesson5 = input.distinct.sorted(Ordering[Int].reverse)(1)
```

**Câu 6 và Câu 7 sử dụng danh sách dưới đây làm input:**

```scala
val input_string = List("apple", "banana", "apricot", "cherry", "blueberry")
```

Câu 6: Hãy biến đổi để thành List((String, Int)) trong đó: 
String là tên ký tự, Int là độ dài của ký tự. 
Kết quả mong muốn: 

_List(("apple", 5), ("banana", 6), ...)_

```scala
val lesson6 = input_string.map(w => (w, w.length))
```

Câu 7: Hãy tìm từ dài nhất có chứa ký tự 'a'

```scala
val lesson7 = input_string.filter(_.contains('a')).map(w => (w, w.length)).maxBy(_._2)
```
