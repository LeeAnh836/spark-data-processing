Các bạn nhập đoạn văn bản vào Databricks theo hướng dẫn trong Lab trước khi thực hiện bài tập nhé:

Kiểm tra xem file đã có trên DBFS hay chưa, nếu như chạy đoạn này bị lỗi, tức là file chưa có trên DBFS,bạn kiểm tra lại đường dẫn và hướng dẫn upload file nhé.

```scala
val path = "dbfs:/FileStore/data_input_lecture3.txt"

display(dbutils.fs.ls(path))
```

Tại Databricks, chúng ta sẽ sử dụng hàm `head()` build-in sẵn để đọc file từ DBFS

```scala
val data = dbutils.fs.head(path)
```

Câu 1: Mỗi 1 đoạn văn có bao nhiêu từ (đánh số đoạn văn bản bắt đầu từ 1)

```scala
val lesson1 = data.split("\\n\\s*\\n")
              .map(_.trim).filter(_.nonEmpty)
              .zipWithIndex
              .map{case (p, i) => (i + 1, p.split("\\s+").length)}
```

Câu 2: Loại bỏ tất cả các từ trong văn bản theo danh sách sau đây: `is`, `are`, `and`, `the`, `a`, `an`, `in`

```scala
val listt = Set("is", "are", "and", "the", "a", "an", "in")
val lesson2 = data.split("\\s+").filterNot(listt.contains).mkString(" ")
```

Câu 3: Thực hiện tính tần suất xuất hiện của từ trong cả văn bản. Trả ra top 10 từ có tần suất xuất hiện nhiều nhất. Sử dụng kết quả của câu 2 để làm tiếp câu 3 (sử dụng `val lesson2`)

```scala
val top10 =
  lesson2
    .toLowerCase
    .replaceAll("[^a-zà-ỹ\\s]", "")   // bỏ dấu câu (khuyên dùng)
    .split("\\s+")
    .filter(_.nonEmpty)
    .groupBy(identity)
    .map { case (word, list) => (word, list.length) }
    .toSeq
    .sortBy { case (_, count) => -count }
    .take(10)
```

Câu 4: Tại mỗi đoạn văn, trả ra từ nào có tần số xuất hiện nhiều nhất

```scala
val result =
  data
    .split("\\n\\s*\\n")                    // tách đoạn văn
    .zipWithIndex
    .map { case (p, idx) =>
      val mostFreq =
        p.toLowerCase
         .replaceAll("[^a-zA-Zà-ỹÀ-Ỹ\\s]", "")
         .split("\\s+")
         .filter(_.nonEmpty)
         .groupBy(identity)
         .mapValues(_.length)
         .maxBy(_._2)

      s"Đoạn ${idx + 1}: '${mostFreq._1}' xuất hiện ${mostFreq._2} lần"
    }

```
